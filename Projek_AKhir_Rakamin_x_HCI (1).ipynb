# Business Understanding


## Problem Statement
Skor kredit adalah angka yang menunjukkan seberapa baik seseorang membayar hutang mereka. Skor ini berkisar antara 300 hingga 850, dengan skor yang lebih tinggi menunjukkan kreditur yang lebih baik.

Skor kredit dipengaruhi oleh beberapa faktor, termasuk:

- Riwayat pembayaran: Kreditur akan melihat bagaimana Anda telah membayar utang Anda di masa lalu.
- Total hutang: Kreditur akan melihat berapa banyak hutang yang Anda miliki - secara keseluruhan.
- Jenis pinjaman: Kreditur akan melihat jenis pinjaman apa yang Anda miliki.
- Panjang riwayat kredit: Kreditur akan melihat berapa lama Anda telah memiliki riwayat kredit.

Skor kredit penting karena dapat memengaruhi kemampuan Anda untuk mendapatkan pinjaman. Kreditur yang memiliki skor kredit yang lebih tinggi lebih cenderung disetujui untuk pinjaman dengan suku bunga yang lebih rendah.

## Busniess Metric dan Objectives

Pemberi pinjaman menggunakan skor kredit untuk mengevaluasi kemungkinan bahwa seseorang akan membayar pinjaman secara tepat waktu. Faktor-faktor yang dipertimbangkan dalam penilaian kredit ini dapat digunakan untuk membangun model untuk memprediksi kemampuan setiap pemohon dalam melunasi pinjaman (skor kredit).

Model prediksi skor kredit dapat dievaluasi dengan menggunakan beberapa metrik, termasuk:

- Akurasi: Persentase kasus yang diprediksi dengan benar oleh model.
- Presisi: Persentase kasus yang benar-benar positif yang diprediksi sebagai positif oleh model.
- Recall: Persentase kasus positif yang benar-benar diprediksi sebagai positif oleh model.
- F1-score: Rata-rata harmonik dari akurasi dan presisi.
- AUC: Area di bawah kurva ROC.

## Model Definition
Model yang dibuat adalah untuk memprediksi skor kredit menggunakan kumpulan data yang disediakan oleh Home Credit Indonesia. Kumpulan data ini terdiri dari 8 data utama, seperti:

- application_train: Data pelatihan untuk model.
- application_test: Data uji untuk model.
- bureau: Data kredit dari biro kredit.
- bureau_balance: Data saldo kredit dari biro kredit.
- previous_application: Data aplikasi kredit sebelumnya.
- OS_CASH_balance: Data saldo POS/tunai.
- installments_payments: Data pembayaran angsuran.
- credit_card_balance: Data saldo kartu kredit.
